```python
a)Plot the graphs of sin x, cos x, e**x and x**2 in [0, 5] in one figure with (2 * 2) subplot
```

```python
from pylab import*
import numpy as np
x=np.linspace(0,5,1000)
y1=sin(x)
y2=cos(x)
y3=e**x
y4=x**2
subplot(2,2,1)
plot(x,y1,label='$sinx$')
legend()
subplot(2,2,2)
plot(x,y2,label='$cosx$')
legend()
subplot(2,2,3)
plot(x,y3,label='$e^x$')
legend()
subplot(2,2,4)
plot(x,y4,label='$x^2$')
legend()
show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return sin(x) + cos (y)
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
X,Y = np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel('y')
title('Projection 3D')
legend()
show()
```

```python
i)write a Python program to solve the following LPP
Max Z = 150x + 75y
Subjected to
4x + 6y <= 24
5x + 3y <= 15
x>0,y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(4*x+6*y<=24)
model +=(5*x+3*y<=15)
model +=150*x+75*y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp module
and simplex method. Find its optimal solution if exist.
Min Z = 3x + 5y + 4z
subject to
2x + 3y <= 8
2y + 5z <= 10
3x + 2y + 5z <= 15
x=>0, y=>0
```

```python
from pulp import*
model=LpProblem(name="Solution",sense=LpMinimize)
x1=LpVariable(name="x1",lowBound=0)
x2=LpVariable(name="x2",lowBound=0)
x3=LpVariable(name="x3",lowBound=0)
model +=(2*x1+3*x2<=24)
model +=(2*x2+5*x3<=15)
model +=(3*x1+2*x2+5*x3<=15)
model +=3*x1+5*x2+4*x3
print(model)
print("\n")
print(model.solve())
print("\n Model Objective value = ",model.objective.value)
print("\nOptimul value Of x1 = ",x1.value())
print("\nOptimul value Of x2 = ",x2.value())
print("\nOptimul value Of x3= ",x3.value())
```
