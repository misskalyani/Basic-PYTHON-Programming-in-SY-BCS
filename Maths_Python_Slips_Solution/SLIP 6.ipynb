```python
(b) Using python, generate 3D surface Plot for the function f (x) = sin (x**2 + y**2 ) in the interval [0, 10].
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return sin (x**2 + y**2 )
x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
X,Y=np.meshgrid(x,y)
Z= f(X,Y)
ax=axes(projection='3d')
ax.plot_surface(X,Y,Z)
xlabel('x')
ylabel('y')
title('3D projection = sin (x**2 + y**2 )')
show()
```

```python
(c)Using Python, plot the graph of function f(x) = sin(x)  e**x + 3*x**2 - log 10(x) on the Interval [0, 2*pi]
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import *
def f(x,y):
    return sin(x)-e**x + 3*x**2 - log (x)
x=np.linspace(0,2*pi,100)
y=np.linspace(0,2*pi,100)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel('y')
title('3D projection =  sin(x)  e**x + 3*x**2 - log 10(x)')
legend()
show()
```

```python
i)write a Python program to solve the following LPP
    Max Z = x + y
    Subjected to
        2x - 2y >=1
        x + y >= 2
        x>0,y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(2*x-2*y>=1)
model +=(x+y>=2)
model +=x+y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp module
   and simplex method. Find its optimal solution if exist.
    Min Z = x + y
    subject to
        x >= 6
        y >= 6
        x + y <= 11
        x>= 0, y>= 0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x>=6)
model +=(y>=6)
model +=(x+y<=11)
model +=x+y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```
