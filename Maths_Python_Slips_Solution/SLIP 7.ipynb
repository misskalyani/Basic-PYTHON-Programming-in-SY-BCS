```python
b)Using python, generate 3D surface Plot for the function f(x) = sin (x**2 + y**2)
in the interval [0,10]
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return np.sin(x**2+y**2)
x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel('y')
legend()
show()
```

```python
i)write a Python program to solve the following LPP
Max Z = 3.5x +2 y
Subjected tox + y >= 5
x >= 4
y <= 2
x>0,y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x<=4)
model +=(y<=15)
model +=3.5*x+2*y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp module
and simplex method. Find its optimal solution if exist.
Min Z = x + 2y + z
subject to
x + 2y + 2z <= 1
3x + 2y + z >= 8
x + y <= 11
x>= 0, y>= 0 , z>=0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x+2*y+2*z<=24)
model +=(5*x+3*y<=15)
model +=150*x+75*y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```
