```python
b)Using Python, plot the surface plot of parabola z = x**2 + y**2 in -6<x,y<6
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return (x**2 + y**2 )
x=np.linspace(-6,6,100)
y=np.linspace(-6,6,100)
X,Y=np.meshgrid(x,y)
Z= f(X,Y)
ax=axes(projection='3d')
ax.plot_surface(X,Y,Z)
xlabel('x')
ylabel('y')
title('3D projection = x**2 + y**2 ')
show()
```

```python
i)write a Python program to solve the following LPP
Max Z = 5x + 3y
Subjected to
x + y <= 7
2x + 5y <= 1
x>0
y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x + y >= 7)
model +=(2*x +5*y >= 1)
model +=5*x + 3*y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp
module and simplex method. Find its optimal solution if exist.
Min Z = 3x +2y +5z
subject to
x + 2y + z <= 430
3x + 2z <= 460
x + 4y <= 120
x=>0, y=>0,z=>0
```

```python
from pulp import*
model=LpProblem(name="Solution",sense=LpMinimize)
x1=LpVariable(name="x1",lowBound=0)
x2=LpVariable(name="x2",lowBound=0)
x3=LpVariable(name="x3",lowBound=0)
model +=( x1+2*x2+2*x3 <= 430)
model +=( 3*x1+4*x3<=460)
model +=(x1+4*x2<=120)
model +=3*x1 +2*x2+5*x3
print(model)
print("\n")
print(model.solve())
print("\n Model Objective value = ",model.objective.value)
print("\nOptimul value Of x1 = ",x1.value())
print("\nOptimul value Of x2 = ",x2.value())
print("\nOptimul value Of x3= ",x3.value())
```

```python
ii)Write a python program to Plot 2D X-axis and Y-axis in black color. In the
same diagram plot:-
(I)Green Triangle with vertices [5,4],[7,4],[6,6]
(II) Blue rectangle with vertices [2, 2], [10, 2], [10, 8], [2, 8].
(III) Red polygon with vertices [6, 2], [10, 4], [8, 7], [4, 8], [2, 4].
(IV) Isosceles triangle with vertices [0, 0], [4, 0], [2, 4].
```

```python
(I)Green Triangle with vertices [5,4],[7,4],[6,6]
```

```python
import matplotlib.pyplot as plt
vertices=[[5,4],[7,4],[6,6]]
x=[vertex[0]for vertex in vertices]
y=[vertex[1]for vertex in vertices]
plt.plot(x+[x[0]],y+[y[0]],color="green")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("black axis with green Triangle")
plt.show()
```
