```python
i)write a Python program to solve the following LPP
Max Z = 3.5x + 2y
Subjected to
x + y >= 5
x >=4
y<=5
x >= 0,y>= 0.
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x+y >= 5)
model +=(x >= 4)
model +=(y<=5)
model +=3.5*x + 2*y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp module
and simplex method. Find its optimal solution if exist.
Min Z = x+y
subject to
x => 6
y => 6
x + y <= 11
x=>0, y=>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x >= 6)
model +=(y >= 6)
model +=(x+y<=11)
model +=x + y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
i)Write the python program to apply each of the following transformation on
the point P(-2,4)
(I)Reflection Through the line y = x+1
(II) Scaling in y-Coordinate by factor 1.5
(III) Shearing in x – Direction by 2 unit
(IV) Rotation about origin by an angle 45 degree.
```

```python
from sympy import*
p=Point(-2,4)
x,y=symbols('x y')
print('\nReflection Through the line y = x+1 = ',p.reflect(Line(x+1)))

print('\nScaling in y-Coordinate by factor 1.5 = ')
p.scale(0,1.5)
print('\n Shearing in x – Direction by 2 unit = ',p.transform(Matrix([[1,0,0],[2,1,0],[0,0,1]])))
print('\nRotation about origin by an angle 45 degree. = ',p.rotate(45))
```
