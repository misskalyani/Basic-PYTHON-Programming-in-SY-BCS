```python
(a)Using Python plot the surface plot of function z = cos (x**2 + y**2 - 0.5) in the interval from -1 < x,y < 1.
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import *
def f(x,y):
    return cos(x**2 + y**2 - 0.5)
x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel('y')
title('3D Surface Plot ')
legend()
show()

```

```python
i)write a Python program to solve the following LPP
    Max Z = 5x + 3y
    Subject to
        x + y <= 7
        2x + 5y <= 1
        x>0
        y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(x+y<=7)
model +=(2*x+5*y<=1)
model +=5*x+3*y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp
module and simplex method. Find its optimal solution if exist.
    Min Z = 4x+y+3z+5w
    subject to
        4x+6y-5z-4w >= 10
        -8x-3y+3z+2w <= 20
        x + y <= 11
        x >= 0,y>= 0,z>= 0,w>= 0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
z=LpVariable(name="z",lowBound=0)
w=LpVariable(name="w",lowBound=0)
model +=(4*x+-6*y-4*w >= -20)
model +=(-8*x-3*y+3*z+2*w <= 20)
model += 4*x+y+3*z+5*w
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
print("\n Optimul Value Of z = ",z.value())
print("\n Optimul Value Of w = ",w.value())
```

```python
Write a python program to Plot 2D X-axis and Y-axis in black color. In the
same diagram plot:-
(I)Green Triangle with vertices [5,4],[7,4],[6,6]
(II) Blue rectangle with vertices [2, 2], [10, 2], [10, 8], [2, 8].
(III) Red polygon with vertices [6, 2], [10, 4], [8, 7], [4, 8], [2, 4].
(IV) Isosceles triangle with vertices [0, 0], [4, 0], [2, 4].
```

```python
(I)Green Triangle with vertices [5,4],[7,4],[6,6]
```

```python
import matplotlib.pyplot as plt
vertices=[[5,4],[7,4],[6,6]]
x=[vertex[0]for vertex in vertices]
y=[vertex[1]for vertex in vertices]
plt.plot(x+[x[0]],y+[y[0]],color="green")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("black axis with green Triangle")
plt.show()
```
