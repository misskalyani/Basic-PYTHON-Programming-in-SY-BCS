```python
(c)Using Python plot the surface plot of parabola z = x**2 + y**2 in -6 <x,y<6
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return x**2 + y**2
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
ax = axes(projection='3d')
ax.contour3D(X,Y,Z,50)
xlabel('x')
ylabel('y')
title('3D Parabola')
legend()
show()
```

```python
(c)Write a Python program to draw a polygon with vertices (0,0), (2,0), (2,3) and (1,6) and rotate by 1800
```

```python
from matplotlib.pylab import*
from sympy import *
a=Point(0,0)
b=Point(2,0)
c=Point(2,3)
d=Point(1,6)
p=Polygon(a,b,c,d)
p.rotate(pi)
```

```python

from matplotlib.patches import Polygon as PolygonPatch
import matplotlib.pyplot as plt
from sympy import Point, Polygon, pi
op = Polygon(Point(0, 0), Point(2, 0), Point(2, 3), Point(1, 6))
rotated = op.rotate(pi)
x_op, y_op = zip(*op.vertices)
x_rotated, y_rotated = zip(*rotated.vertices)
plt.gca().add_patch(PolygonPatch(list(zip(x_op, y_op)), alpha=0.3, label='Original', color='red'))
plt.gca().add_patch(PolygonPatch(list(zip(x_rotated, y_rotated)), alpha=0.3, linestyle='--', label='Rot'))
plt.axis('equal')
plt.legend()
plt.show()
```

```python
(i) Write a Python program to solve the following LPP:
    Max Z = 150*x+75*y
    subject to
             4*x+6*y<=24
             5*x+3*y<=15
            x ≥ 0, y ≥ 0.
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(4*x+6*y<=24)
model +=(5*x+3*y<=15)
model +=150*x+75*y
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
```

```python
(ii)) Write a python program to display the following LPP by using pulp module and simplex method. 
    Find its optimal solution if exist.
    Min Z = 4x+y+3z+5w
    subject to
            4x+-6y-4w >= -20
            -8x-3y+3z+2w <= 20
            x + y <= 11
            x >= 0,y>= 0,z>= 0,w>= 0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
z=LpVariable(name="z",lowBound=0)
w=LpVariable(name="w",lowBound=0)
model +=(4*x+-6*y-4*w >= -20)
model +=(-8*x-3*y+3*z+2*w <= 20)
model += 4*x+y+3*z+5*w
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
print("\n Optimul Value Of z = ",z.value())
print("\n Optimul Value Of w = ",w.value())
```
