```python
i)write a Python program to solve the following LPP
Max Z = 150x + 75y
Subjected to
4x + 6y <= 24
5x + 3y <= 15
x>0,y>0
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(4*x+6*y<=24)
model +=(5*x+3*y<=15)
model +=150*x+75*y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
ii)Write a python program to display the following LPP by using pulp module
and simplex method. Find its optimal solution if exist.
Min Z = 4x + y + 3z + 5w
subject to
4x + 6y – 5z – 4w >= 20
-3x – 2y + 2z + w <= 10
-8x – 3y + 3z + 2w <= 20
x=>0, y=>0,z=> 0,w>=
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
z=LpVariable(name="z",lowBound=0)
w=LpVariable(name="w",lowBound=0)
model +=(4*x+-6*y-4*w >= -20)
model +=(-8*x-3*y+3*z+2*w <= 20)
model += 4*x+y+3*z+5*w
print(model)
print("\n")
print(model.solve())
print("\n model Objective value = ",model.objective.value())
print("\n Optimul Value Of x = ",x.value())
print("\n Optimul Value Of y = ",y.value())
print("\n Optimul Value Of z = ",z.value())
print("\n Optimul Value Of w = ",w.value())
```
