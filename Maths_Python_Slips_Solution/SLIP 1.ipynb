```python
(b) Write a Python program to draw a polygon with vertices (0, 0), (2, 0), (2, 3) and (1, 6) and rotate
it by 180◦ .
```

```python
from matplotlib.pylab import* 
from sympy import *
a=Point(0,0)
b=Point(2,0)
c=Point(2,3)
d=Point(1,6)
p=Polygon(a,b,c,d)
p.rotate(pi) # print(p.rotate(pi)) 
```

```python

from matplotlib.patches import Polygon as PolygonPatch
import matplotlib.pyplot as plt
from sympy import Point, Polygon, pi
op = Polygon(Point(0, 0), Point(2, 0), Point(2, 3), Point(1, 6))
rotated = op.rotate(pi)
x_op, y_op = zip(*op.vertices)
x_rotated, y_rotated = zip(*rotated.vertices)
plt.gca().add_patch(PolygonPatch(list(zip(x_op, y_op)), alpha=0.3, label='Original', color='red'))
plt.gca().add_patch(PolygonPatch(list(zip(x_rotated, y_rotated)), alpha=0.3, linestyle='--', label='Rot'))
plt.axis('equal')
plt.legend()
plt.show()
```

```python
(i) Write a Python program to solve the following LPP:
Max Z = 150x + 75y
subject to
4x + 6y ≤ 24
5x + 3y ≤ 15
x ≥ 0, y ≥ 0.
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMaximize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model +=(4*x+6*y<=24)
model +=(5*x+3*y<=15)
model +=150*x+75*y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```

```python
(ii) Write a python program to display the following LPP by using pulp module and simplex
method. Find its optimal solution if exist.
Min Z = x + y
subject to
    x ≥ 6
    y ≥ 6
    x + y ≤ 11
    x ≥ 0, y ≥ 0.
```

```python
from pulp import *
model=LpProblem(name="Solution",sense=LpMinimize)
x=LpVariable(name="x",lowBound=0)
y=LpVariable(name="y",lowBound=0)
model += (x>=6)
model += (y>=6)
model += (x+y<=11)
model += x+y
print(model)
print('\n')
print(model.solve())
print('\nModel Objective Value = ',model.objective.value())
print('\nValue of x = ',x.value())
print('\nValue of y = ',y.value())
```
